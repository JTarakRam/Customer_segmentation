### Annual Income

```python
income_df = data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
income_df = ss.fit_transform(income_df)
```

```python
# Determine the optimal number of clusters using the elbow method
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(income_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# we can see that 5 clusters are suitable for the clustering.
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(income_df)
y_kmeans
```

```python
# Visualize the resulting clusters
plt.figure(figsize=(16,8))
plt.scatter(income_df[y_kmeans == 0, 0], income_df[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(income_df[y_kmeans == 1, 0], income_df[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(income_df[y_kmeans == 2, 0], income_df[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(income_df[y_kmeans == 3, 0], income_df[y_kmeans == 3, 1], s = 100, c = 'brown', label = 'Cluster 4')
plt.scatter(income_df[y_kmeans == 4, 0], income_df[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

Cluster 1 (red): Customers in this cluster have a low annual income and a low spending score. These customers may be price-sensitive and tend to prioritize practical purchases over luxury or indulgent items.

Cluster 2 (blue): Customers in this cluster have a moderate annual income and a moderate spending score. These customers may represent a "typical" consumer who shops for a variety of goods and services.

Cluster 3 (green): Customers in this cluster have a high annual income but a low spending score. These customers may be conservative or practical in their purchasing habits and may prioritize saving or investing their income rather than spending it.

Cluster 4 (brown): Customers in this cluster have a low annual income but a high spending score. These customers may be young or inexperienced shoppers who prioritize immediate gratification over long-term financial planning.

Cluster 5 (magenta): Customers in this cluster have a high annual income and a high spending score. These customers may be affluent or luxury consumers who prioritize premium products and experiences.
